```python
#Renaming the columns in the datasets
df = df.rename(columns = {'name': 'Player_Name', 'shirt_number': 'Shirt_Number', 'team_name': 'Team_Name', 'height': 'Height',
                       'weight': 'Weight', 'age': 'Age', 'foot': 'Strong_Foot', 'registered_position': 'Player_Position',
                       'offensive_awareness': 'Attack_Awareness', 'ball_control': 'Ball_Control', 'dribbling': 'Dribbling',
                       'low_pass': 'Low_Pass', 'lofted_pass': 'Lofted_Pass', 'finishing': 'Finish', 'heading': 'Header',
                       'place_kicking': 'Freekick', 'curl': 'Curl', 'speed': 'Speed', 'acceleration': 'Acceleration',
                       'kicking_power': 'Shot_Power', 'jump': 'Jump_Power', 'stamina': 'Stamina',
                       'defensive_awareness': 'Defense', 'ball_winning': 'Ball_Winning', 'aggression': 'Aggression',
                       'weak_foot_accuracy':'Weak_Foot', 'form':'Form', 'injury_resistance':'Injury_Resistance',
                        'overall_rating':'Player_Rating'
                      }, 
               inplace = False)
```

```python
df.head()
```

```python
#dropping column in our dataset
cl = ['condition', 'physical_contact', 'balance']
df =df.drop(cl, axis=1, inplace = False)
df.sample(5).T
```

```python
#Fastest players
fastest_players =df[["Acceleration","Player_Name","Player_Position",
                     'Age','Team_Name','Speed',"Player_Rating"]].nlargest(5, ['Acceleration']).set_index('Player_Name')
fastest_players
```

```python
#Tallest players
tallest_players =df[["Height","Player_Name","Player_Position",
                     'Age','Team_Name', 'Weight',"Player_Rating"]].nlargest(5, ['Height']).set_index('Player_Name')
tallest_players
```

```python
#strongest players
strongest_players = df[["Stamina","Player_Name","Player_Position",
                     'Age','Team_Name', 'Weight',"Player_Rating"]].nlargest(5, ['Stamina']).set_index('Player_Name')
strongest_players
```

```python
#lofted pass
best_loft_passes = df[["Lofted_Pass","Player_Name","Player_Position",
                     'Age','Team_Name',"Player_Rating"]].nlargest(5, ['Lofted_Pass']).set_index('Player_Name')
best_loft_passes
```

```python
#low pass
best_low_passes = df[["Low_Pass","Player_Name","Player_Position",
                     'Age','Team_Name',"Player_Rating"]].nlargest(5, ['Low_Pass']).set_index('Player_Name')
best_low_passes
```

```python
#best defenders
best_defenders = df[["Defense","Player_Name","Player_Position",
                     'Age','Team_Name',"Player_Rating"]].nlargest(5, ['Defense']).set_index('Player_Name')
best_defenders
```

```python
#best defenders
best_attackers = df[["Attack_Awareness","Player_Name","Player_Position",
                     'Age','Team_Name',"Player_Rating"]].nlargest(5, ['Attack_Awareness']).set_index('Player_Name')
best_attackers
```

```python
#olldest player
oldest_players = df[["Age","Player_Name","Player_Position",
                    'Team_Name',"Player_Rating"]].nlargest(5, ['Age']).set_index('Player_Name')
oldest_players
```

```python
#youngest players
youngest_players = df[["Age","Player_Name","Player_Position",
                    'Team_Name',"Player_Rating"]].nsmallest(5, ['Age']).set_index('Player_Name')
youngest_players
```

```python
#top dribbler
top_dribbler = df[["Dribbling","Age","Player_Name","Player_Position",
                    'Team_Name',"Player_Rating"]].nlargest(1, ['Dribbling']).set_index('Player_Name')
top_dribbler
```

```python
#top finisher
top_finisher = df[["Finish","Age","Player_Name","Player_Position",
                    'Team_Name',"Player_Rating"]].nlargest(1, ['Finish']).set_index('Player_Name')
top_finisher
```
